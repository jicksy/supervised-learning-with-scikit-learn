```python
# Aim: To perform Exploratory Data Analysis (EDA) using Scatter Plot on Iris dataset
# import necessary methods

from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# loading dataset
iris = datasets.load_iris()
```

```python
# checking  type of iris
type(iris)
```

```python
# Bunch is similar to dictionary with key-value pairs. Let's print keys.
print(iris.keys())
```

```python
# both are numpy arrays
type(iris.data), type(iris.target)
```

```python
# Let's get the shape of data with command iris.data.shape 
# 150 rows and 4 columns, i.e., 150 samples and 4 features (petal length, petal width, sepal length, sepal width.
iris.data.shape
```

```python
# Target names; sertosa - index(0), versicolor - index(1), virginica - index(2)
iris.target_names
```

```python
# To perform Exploratory data analysis

# Assign feature data to X
X = iris.data

# Assign target data to y
y = iris.target
```

```python
# build DataFrame
df = pd.DataFrame(X, columns = iris.feature_names)
```

```python
# df.head() shows us the first five rows of data
print(df.head())
```

```python
# Visual EDA; Let's now plot the information as a Scatter Plot. 
# A scatter plot displays the correlation between a pair of variables. 
# Given a set of n variables, there are n-choose-2 pairs of variables, and thus the same numbers of scatter plots.

_ = pd.plotting.scatter_matrix(df, c = y, figsize = [8, 8], s = 150, marker = 'D')
```
